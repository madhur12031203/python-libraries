## Numpy array and basics

```python
import numpy as np
```

Creating list from array

```python
arr1=np.array([1,2,3,4,5])
print("1D ARRAY:",arr1)

arr2=np.array([[1,2,3],[4,5,6]])
print("2D ARRAY:",arr2)
```

#List vs numpy array

```python
py_list=[1,2,3]
print("Multiplication of list",py_list*2)

np_array=np.array([1,2,3])
print("Multiplication of numpy array",np_array*2) # 2 is scalar


import time
start=time.time()
py_list=[i*2 for i in range(100000)]
print("\n List time:",time.time()-start)

start=time.time()
np_arry=np.arange(100000)*2
print("\n numpy time:",time.time()-start)
```

creating array from scratch

```python
zeros=np.zeros((3,4))
print(zeros)

ones=np.ones((3,4))
print(ones)

full=np.full((2,2),6)
print(full)

random=np.random.randint((2,3))
print(random)

seq=np.arange(0,11,2)
print(seq)
```

## vector , matrice and tensor


```python
vector=np.array([1,2,3])
print(vector)

matrix=np.array([[1,2,3],[4,5,6]])
print(matrix)

tensor=np.array([[[1,2],[3,4]],
                 [[5,6],[7,8]]])
print(tensor)
```

# properties of array

```python
arr=np.array([[1,2,3],[4,5,True]])
print("Shape",arr.shape)
print("Dimensions",arr.ndim)
print("Size",arr.size)
print("Dtype",arr.dtype)
```

### Array Reshaping

```python
arr=np.arange(12)
print(arr)

reshaped=arr.reshape((2,6))
print(reshaped)

flattened=reshaped.flatten()
print(flattened)

raveled=reshaped.ravel()
print(raveled)

transpose=reshaped.T
print(transpose)
```
